```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Carregar o conjunto de dados Iris
iris = load_iris()
X = iris.data
y = iris.target
features = iris.feature_names

# Padronizar os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Aplicar PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Explicação da variância
explained_variance = pca.explained_variance_ratio_
```

```python
# Criar um DataFrame para a variância explicada
variance_df = pd.DataFrame({
    'Componente': [f'PC{i+1}' for i in range(len(explained_variance))],
    'Variância Explicada': explained_variance
})

# Visualizar a variância explicada
print(variance_df)

# Gráfico da variância explicada
plt.figure(figsize=(8, 5))
plt.bar(variance_df['Componente'], variance_df['Variância Explicada'], color='skyblue')
plt.xlabel('Componentes Principais')
plt.ylabel('Proporção da Variância Explicada')
plt.title('Variância Explicada por Cada Componente Principal')
plt.grid(axis='y')
plt.show()
```
