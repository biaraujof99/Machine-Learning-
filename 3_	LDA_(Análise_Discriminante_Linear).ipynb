```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
# Carregar o conjunto de dados Iris
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Inicializar o LDA
lda = LDA(n_components=2)

# Ajustar e transformar os dados
X_lda = lda.fit_transform(X, y)
```

```python

print("Coeficientes dos componentes LDA:")
print(lda.coef_)
```

```python
# Criar um DataFrame com os dados reduzidos
df_lda = pd.DataFrame(data=X_lda, columns=['Componente LDA 1', 'Componente LDA 2'])
df_lda['target'] = y
```

```python
# Visualizar os dados transformados
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Componente LDA 1', y='Componente LDA 2', hue='target', data=df_lda, palette='viridis', s=100)
plt.title('Dados Transformados (LDA)')
plt.xlabel('Componente LDA 1')
plt.ylabel('Componente LDA 2')
plt.legend(title='Classe', loc='upper right', labels=iris.target_names)
plt.grid()
plt.show()
```
