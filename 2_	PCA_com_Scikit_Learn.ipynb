```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
# Carregar o conjunto de dados Iris
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Inicializar o PCA para 2 componentes principais
pca = PCA(n_components=2)

# Ajustar e transformar os dados
X_pca = pca.fit_transform(X)
```

```python
# Vari창ncia explicada
explained_variance = pca.explained_variance_ratio_

# Exibir a vari창ncia explicada
print("Vari창ncia explicada pelos componentes principais:")
print(explained_variance)
```

```python
# Criar um DataFrame com os dados reduzidos
df_pca = pd.DataFrame(data=X_pca, columns=['Componente Principal 1', 'Componente Principal 2'])
df_pca['target'] = y

# Visualizar os dados transformados
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Componente Principal 1', y='Componente Principal 2', hue='target', data=df_pca, palette='viridis', s=100)
plt.title('Dados Transformados (PCA)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(title='Classe', loc='upper right', labels=iris.target_names)
plt.grid()
plt.show()
```
